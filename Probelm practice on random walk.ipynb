```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_1d_absorbing_sites():
    # Initialize the particle's starting position at 0
    position = 0
    # Initialize the number of steps taken
    steps = 0
    
    # Absorbing sites at 10 and -10
    absorbing_sites = [10, -10]
    
    # Perform the random walk until the particle reaches an absorbing site
    while position not in absorbing_sites:
        # Choose a random direction: -1 for left, 1 for right
        direction = np.random.choice([-1, 1])
        # Update the position based on the chosen direction
        position += direction
        # Increment the step count
        steps += 1
    
    # Return the number of steps taken to reach an absorbing site
    return steps

# Number of simulations
num_simulations = 1000

# List to store the results
steps_list = []

# Run the simulations
for _ in range(num_simulations):
    steps = random_walk_1d_absorbing_sites()
    steps_list.append(steps)

# Calculate the average number of steps taken to reach an absorbing site
average_steps = np.mean(steps_list)

# Print the average number of steps
print(f"Average number of steps taken to reach an absorbing site: {average_steps:.2f}")

# Plot a histogram of the number of steps taken in each simulation
plt.hist(steps_list, bins=20, alpha=0.7, color='b')
plt.title('Histogram of Steps Taken to Reach Absorbing Sites')
plt.xlabel('Number of Steps')
plt.ylabel('Frequency')
plt.show()
```
